*ETL Process*<br>
Extract, transform, load (ETL) is the general procedure of copying data from one or more sources into a destination system which represents the data differently from the source(s) or in a different context than the source(s). The ETL process became a popular concept in the 1970s and is often used in data warehousing.

## Comparing price by room types

We first filter columns to get details of each neighborhood group and room type,
then index price for that specific room type in specific neighborhood. (pandas .loc function to index rows)

### Shared rooms

```python
shared_brook=nyc_pd.loc[(nyc_pd['neighbourhood_group'] == 'Brooklyn') & (nyc_pd['room_type']=='Shared room')]
price_sub1=num_airbnb['price'].iloc[shared_brook.index]

shared_bronx=nyc_pd.loc[(nyc_pd['neighbourhood_group'] == 'Bronx') & (nyc_pd['room_type']=='Shared room')]
price_sub2=num_airbnb['price'].iloc[shared_bronx.index]

shared_si=nyc_pd.loc[(nyc_pd['neighbourhood_group'] == 'Staten Island') & (nyc_pd['room_type']=='Shared room')]
price_sub3=num_airbnb['price'].iloc[shared_si.index]

shared_queens=nyc_pd.loc[(nyc_pd['neighbourhood_group'] == 'Queens') & (nyc_pd['room_type']=='Shared room')]
price_sub4=num_airbnb['price'].iloc[shared_queens.index]

shared_mttn=nyc_pd.loc[(nyc_pd['neighbourhood_group'] == 'Manhattan') & (nyc_pd['room_type']=='Shared room')]
price_sub5=num_airbnb['price'].iloc[shared_mttn.index]

brooklyn_perprice=[]
bronx_perprice = []
si_perprice = []
queens_perprice = []
mttn_perprice = []
percent = []
for i in range (25,91):
    brooklyn_perprice.append(int(price_sub1.quantile(i/100)))
    bronx_perprice.append(int(price_sub2.quantile(i/100)))
    si_perprice.append(int(price_sub3.quantile(i/100)))
    queens_perprice.append(int(price_sub4.quantile(i/100)))
    mttn_perprice.append(int(price_sub5.quantile(i/100)))
    percent.append(i)
```

```python
plt.title('Prices of 90% Shared Rooms',fontsize=15)
sd=sns.lineplot(x=pd.Series(percent),y=pd.Series(brooklyn_perprice),label='Brooklyn')
sd=sns.lineplot(x=pd.Series(percent),y=pd.Series(bronx_perprice),label='Bronx')
sd=sns.lineplot(x=pd.Series(percent),y=pd.Series(si_perprice),label='Staten Island')
sd=sns.lineplot(x=pd.Series(percent),y=pd.Series(queens_perprice),label='Queens')
sd=sns.lineplot(x=pd.Series(percent),y=pd.Series(mttn_perprice),label='Manhattan')

sd.set(xlabel='Percentiles', ylabel='Prices in U.S $')
```

Price distribution shows that cheapest in Bronx and the most expensive in Manhattan(90% of shared rooms is near $150)

## Data Preprocessing for Prediction

*We are focusing on price predictions, so to see possible future prices based on data using Machine learning techniques along with Linear Regression models.*

```python
nyc_pd.drop(['name','id','host_name','last_review'], axis=1, inplace=True)
nyc_pd['reviews_per_month']=nyc_pd['reviews_per_month'].replace(np.nan, 0)
```

Encode labels with value 0 and 1 since we cant have texts in our data.

```python
preprocess = preprocessing.LabelEncoder()
preprocess.fit(nyc_pd['neighbourhood_group'])
nyc_pd['neighbourhood_group'] = preprocess.transform(nyc_pd['neighbourhood_group'])

preprocess = preprocessing.LabelEncoder()
preprocess.fit(nyc_pd['neighbourhood'])
nyc_pd['neighbourhood'] = preprocess.transform(nyc_pd['neighbourhood'])

preprocess = preprocessing.LabelEncoder()
preprocess.fit(nyc_pd['room_type'])
nyc_pd['room_type'] = preprocess.transform(nyc_pd['room_type'])

nyc_pd.sort_values(by='price', ascending = True, inplace = True)

nyc_pd.head()
```

```python
list(preprocess.inverse_transform(nyc_pd['room_type']))[:10]
```

### Linear Regression Model

```python
lrm = LinearRegression()
X = nyc_pd[['host_id','neighbourhood_group','neighbourhood','latitude','longitude',\
            'room_type','minimum_nights','number_of_reviews','reviews_per_month',\
            'calculated_host_listings_count','availability_365']]
Y = nyc_pd['price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
lrm.fit(X_train,Y_train)
```

Predictions and Metrics

```python
prediction = lrm.predict(X_test)

print("""
        Mean Squared Error: {}
        R2 Score: {}
        Mean Absolute Error: {}
     """.format(
        np.sqrt(metrics.mean_squared_error(Y_test, prediction)),
        r2_score(Y_test,prediction) * 100,
        mean_absolute_error(Y_test,prediction)
        ))
```

Results for Actual and Predicted Values

```python
error_data = pd.DataFrame({"Actual Values": np.array(Y_test).flatten(),
                           "Predicted Values":prediction.flatten()}).head(20)
error_data.head()
```

```python
plt.figure(figsize=(16,8))
sns.regplot(prediction,Y_test)
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.title("Linear Model Predictions")
plt.show()
```
